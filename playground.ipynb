```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
import csv 

f_without_bout = open('no_blankout.log')
f_with_bout = open('blankout.log')
reader1 = csv.reader(f_without_bout, delimiter=',')
reader2 = csv.reader(f_with_bout, delimiter=',')

bout = []
no_bout = []
num = 0
num2 = 0
for row1 in reader1:
    if num < 1000:
        no_bout.append(row1)
    num += 1

num2 = 0
for row2 in reader2:
    if num2 < 1000:
        bout.append(row2)
    num2 += 1

```

```python
# into plot format 
def loss(l):
    return map(lambda (step, loss, m, v, t) : loss, l)
def minibatch(l):
    return map(lambda (step, loss, m, v, t) : float(m.rstrip('%% ').lstrip(' ')), l)
def val_acc(l):
    return map(lambda (step, loss, m, v, t) : v.rstrip('%%'), l)
def test_acc(l):
    return map(lambda (step, loss, m, v, t) : t.rstrip('%% '), l)

plt.figure(figsize=(20,10))
plt.ylim(0,100)
print len(loss(bout)), len(minibatch(bout))
plt.plot(loss(bout))
plt.plot(minibatch(bout))
plt.plot(val_acc(bout))
plt.plot(test_acc(bout))
```
